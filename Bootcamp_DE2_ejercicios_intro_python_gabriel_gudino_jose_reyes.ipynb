## 2) Preparacion de datos

2.a) Obtener nulos de la columna species (tip: seleccionar primero los nulos de la columna, por ejemplo 'dataframe['columna'].isna()' y usarlos luego para filtrar los nulos)

```python
nulos_species = starwars["species"][starwars["species"].isna() == True]
print(nulos_species)
```

2.b) Reemplazar nulos en la columna 'species' por el string 'Unknown'

```python
# starwars["species"][starwars["species"].isna()] = "Unknown"
starwars["species"] = starwars["species"].fillna(value="Unknown")
```

2.c) Reemplazar nulos de la columna 'mass' por la mediana de la columna

```python
# starwars["mass"][starwars["mass"].isna()] = starwars["mass"].mean()
starwars["mass"] = starwars["mass"].fillna(value= starwars["mass"].median())
```

2.d) Eliminar las columnas 'sex' y 'skin_color'

```python
starwars = starwars.drop(columns = ["sex", "skin_color"])
```

2.f) Ordenar el dataframe ascendentemente por la columna mass

```python
starwars = starwars.sort_values("mass", ignore_index= True)
```

2.g) Crear una nueva columna llamada mass_lb

```python
starwars["mass_lb"] = starwars["mass"] * 2.2
```

2.h) Renombrar la columna 'mass' a 'mass_kg'

```python
starwars = starwars.rename(columns = {"mass":"mass_kg"})
```

2.i) Reordenar las columnas del dataframe, ubicando 'mass_lb' a la derecha de 'mass_kg'

```python
starwars = starwars[["name", "height", "mass_kg", "mass_lb", "hair_color", "eye_color", "birth_year", "gender", "homeworld", "species"]]
```

2.j) Eliminar del dataset, los registros en donde mass_kg sea menor a 20

```python
starwars = starwars[starwars["mass_kg"] >= 20]
```

## 3) Exportar dataframe

Exportar a CSV, usando como separador el caracter |

```python
starwars.to_csv("d_star_wars.csv", sep = "|" )
```
